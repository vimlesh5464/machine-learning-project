```python
# Student Performance EDA
# Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load Dataset
df = pd.read_csv("../data/students.csv")
print(df.head())
print(df.info())

# 2. Clean Data
df = df.drop_duplicates()
df = df.dropna(subset=["Hours_Studied", "Score"])
df["Salary"] = df["Salary"].fillna(df["Salary"].median())
df.to_csv("../outputs/cleaned_data.csv", index=False)

# 3. NumPy Statistics
hours = df["Hours_Studied"].to_numpy()
scores = df["Score"].to_numpy()

print("Mean Hours Studied:", np.mean(hours))
print("Median Hours Studied:", np.median(hours))
print("Std Dev of Scores:", np.std(scores))
print("Correlation (Hours vs Score):", np.corrcoef(hours, scores)[0,1])

# 4. Visualizations

# a) Histogram: Hours Studied
plt.figure(figsize=(6,4))
sns.histplot(df["Hours_Studied"], bins=10, kde=True, color="skyblue")
plt.title("Distribution of Hours Studied")
plt.xlabel("Hours Studied")
plt.ylabel("Count")
plt.savefig("../outputs/plots/hours_histogram.png")
plt.show()

# b) Scatter Plot: Hours vs Score (Pass/Fail)
df["Pass"] = df["Score"] >= 40
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x="Hours_Studied", y="Score", hue="Pass", palette={True:"green", False:"red"})
plt.title("Hours Studied vs Score")
plt.savefig("../outputs/plots/hours_vs_score.png")
plt.show()

# c) Boxplot: Salary
plt.figure(figsize=(6,4))
sns.boxplot(data=df, y="Salary")
plt.title("Salary Distribution")
plt.savefig("../outputs/plots/salary_boxplot.png")
plt.show()

# d) Boxplot: Score by Performance
plt.figure(figsize=(6,4))
sns.boxplot(x="Performance", y="Score", data=df)
plt.title("Score Distribution by Performance")
plt.savefig("../outputs/plots/performance_boxplot.png")
plt.show()

# e) Heatmap: Correlation
plt.figure(figsize=(6,5))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.savefig("../outputs/plots/heatmap.png")
plt.show()
```
